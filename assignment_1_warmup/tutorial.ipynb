````python
import numpy as np
from bisection_method import bisect, plotter_function
````

# Example 1. 
```python
f = lambda x: (x-0.5)**3 
```

````python
# a: lower bound of the interval
a = 0
# b: upper bound of the interval
b = 2
# max_iter: maximum number of iterations
max_iter = 100
f = lambda x: (x-0.5)**3 
c, f_c, c_history = bisect(f, [a, b], max_iter, TOL = 1e-6)
print(f"root = {c}")
# Plotting the results and steps to the root
plotter_function(f, c_history,  "eq3", [-1, 2])
````

# Example 2.

````python
a = 0
## b: upper bound of the interval
b = 2
max_iter = 100
# f: function to find the root
K = 1
l = 1
F = 1.5
a_ = 1
b_ = 10
f = lambda w: 2*K*(np.sqrt(l**2 + w**2) - l) * (w)/(np.sqrt(l**2 + w**2)) - F
c, f_c, c_history = bisect(f, [a, b], max_iter, TOL = 1e-6)
print(f"root = {c}")
# Plotting the results and steps to the root
plotter_function(f, c_history,  "eq4", [-1, 2])
````
